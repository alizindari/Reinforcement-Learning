```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Environment:
    def __init__(self,numOfBandit,initial_qStar,variance):
        self.qStar = initial_qStar
        self.variance = variance
        self.numOfBandit = numOfBandit
        self.meanOfqStar = []
        self.best_reward = []
        
    def step(self,action):
        return np.random.normal(self.qStar[action],self.variance[action])
    
    def change_qStars(self):
        for i in range(len(self.qStar)):
            self.qStar[i]+= np.random.randn()
        self.meanOfqStar.append(np.mean(self.qStar))
        self.best_reward.append(np.max(self.qStar))
```

```python
    
class Agent:
    def __init__(self,environment,epsilon,method,alpha):
        self.reward = []
        self.env = environment
        self.historyOfReward = []
        self.historyOfLast10Rewards =[]
        self.epsilon = epsilon
        self.qValues = []
        self.numOfAction = []
        self.alpha = alpha
        self.numOfBandit = environment.numOfBandit
        self.method = method
        for i in range(self.numOfBandit):
            self.reward.append(0)
            self.qValues.append(0)
            self.numOfAction.append(0)
    
        
     
    def updateQvalues(self,reward,action):
        if self.method == 'const':
            self.qValues[action] += self.alpha*(reward-self.qValues[action])
        else:
            self.qValues[action] += (1/self.numOfAction[action])*(reward-self.qValues[action])
            
        
        
        
    def takeAction(self):
        if np.random.rand()<self.epsilon:
            action = np.argmax(self.qValues)
        else:
            action = np.random.randint(self.numOfBandit)
         
        self.numOfAction[action] += 1
        return action
    

            
        
```

```python
mean = [10,11,20,11,12,10]
variance = [.5,.5,.5,.5,.5,.5]
alpha = 0.9

print(len(mean),len(variance))

env = Environment(6,mean,variance,)
ag1 = Agent(env,0.85,'const',alpha)
ag2 = Agent(env,0.85,'non-const',alpha)

sumOfReward1 = 0
sumOfReward2 = 0

for i in range(20000):
    action = ag1.takeAction()
    reward = env.step(action)
    ag1.updateQvalues(reward,action)
    sumOfReward1 += reward
    
    action = ag2.takeAction()
    reward = env.step(action)
    ag2.updateQvalues(reward,action)
    sumOfReward2 += reward
    
    env.change_qStars()
    
#     print(f'action at level {i} is: {action} , reward : {reward} , mean reward : {np.round(rewardMean,2)}')
    ag1.historyOfReward.append(sumOfReward1/(i+1))
    ag2.historyOfReward.append(sumOfReward2/(i+1))
    
    if i >100:
        ag1.historyOfLast10Rewards.append( (ag1.historyOfReward[i]*i - ag1.historyOfReward[i-100]*(i-10))/100)
        ag2.historyOfLast10Rewards.append( (ag2.historyOfReward[i]*i - ag2.historyOfReward[i-100]*(i-10))/100)

```
