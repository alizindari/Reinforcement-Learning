```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
    
class k_armed_bandit:
    def __init__(self,numOfBandit,mean,variance,epsilon,alpha):
        self.reward = []
        self.historyOfReward = []
        self.epsilon = epsilon
        self.qValues = []
        self.numOfAction = []
        self.alpha = alpha
        self.num = numOfBandit 
        self.rewardMean = mean
        self.rewardVariance = variance
        for i in range(numOfBandit):
            self.reward.append([])
            self.qValues.append(0)
            self.numOfAction.append(0)
        
    def generateReward(self,action):
    
        self.reward[action].append(np.random.normal(self.rewardMean[action],self.rewardVariance[action]))
        return self.reward[action][-1]
    
    
    def __computeNewQvalue(self,action,alpha):
        
        res = 0
        betha = 1-alpha
        for i in range(self.numOfAction[action]):
            res += alpha*(np.power(betha,self.numOfAction[action]-i))*self.reward[action][i]
        return res
    
     
    def updateQvalues(self,action):
        self.qValues[action] = self.__computeNewQvalue(action,self.alpha)
        
        
    def takeAction(self,epsilon):
        if np.random.rand()<epsilon:
            action = np.argmax(self.qValues)
        else:
            action = np.random.randint(self.num)
            
        return action
    
    def main(self,eps): 
        meanReward = 0
        for i in range(eps):
            action = self.takeAction(self.epsilon)
            print(f'action at level {i} is: {action}')
            reward = self.generateReward(action)
            meanReward += reward
            print(f'total reward {meanReward} ')
            self.historyOfReward.append(meanReward/(i+1))
            self.updateQvalues(action)
            self.numOfAction[action] += 1
            
        
```

```python
mean = [1,1,10,1,1,1]
variance = [1,1,1,1,1,1]
alpha = 0.5

print(len(mean),len(variance))

env1 = k_armed_bandit(6,mean,variance,1,alpha)
env2 = k_armed_bandit(6,mean,variance,0.9,alpha)
env3 = k_armed_bandit(6,mean,variance,0.95,alpha)
env4 = k_armed_bandit(6,mean,variance,0.98,alpha)

env1.main(500)
# env2.main(500)
# env3.main(500)
# env4.main(500)
```
